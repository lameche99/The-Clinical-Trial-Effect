```python
import numpy as np
import pandas as pd
```

```python
pd.set_option('display.min_rows', 20)
```

```python
import requests

df = pd.read_csv('../data-src/tickers.csv')
tickers = list(df.columns)
# Your Polygon.io API key
api_key = 'Insert API Key'
df_tot = pd.DataFrame()
# List of tickers
years = ['2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021','2022','2023']
# Iterate over tickers and fetch financial data
for ticker in tickers:
    print(f'Doing {ticker}')
    net_income = []
    revenues = []
    operating_income = []
    total_equity = []
    total_assets = []
    total_liabilities = []
    for year in years:
        url = f'https://api.polygon.io/vX/reference/financials?ticker={ticker}&period_of_report_date.gte={year}-01-01&period_of_report_date.lte={year}-12-31&limit=100&apiKey=0cWQq_K4hIpXGKtVRB80dNvtiSm4OiNJ'
        response = requests.get(url)
        if response.status_code == 200:
            data = response.json()
            results = data.get('results', [])
            if results:
                financials = results[0].get('financials', {})
                balance_sheet = financials.get('balance_sheet', {})
                income_statement = financials.get('income_statement', {})

                # Extract necessary fields
                revenues
                net_income.append(income_statement.get('net_income_loss', {}).get('value', 0))
                operating_income.append(income_statement.get('operating_income_loss', {}).get('value', 0))
                revenues.append(income_statement.get('revenues', {}).get('value', 0))
                
                total_equity.append(balance_sheet.get('equity', {}).get('value', 0))
                total_assets.append(balance_sheet.get('assets', {}).get('value', 0))
                total_liabilities.append(balance_sheet.get('liabilities', {}).get('value', 0))
                
                ### Calculate ratios 
                # ROE
                num = np.array(net_income)
                den = np.array(total_equity)
                out_array = np.zeros_like(num, dtype=np.float64)
                roe = np.divide(num, den, out=out_array, where=(den != 0))
                # Operating Margin
                num = np.array(operating_income)
                den = np.array(revenues)
                out_array = np.zeros_like(num, dtype=np.float64)
                om = np.divide(num, den, out=out_array, where=(den != 0))
                # Net Margin
                num = np.array(net_income)
                den = np.array(revenues)
                out_array = np.zeros_like(num, dtype=np.float64)
                nm = np.divide(num, den, out=out_array, where=(den != 0))
                # Debt Ratio
                num = np.array(total_assets)
                den = np.array(total_liabilities)
                out_array = np.zeros_like(num, dtype=np.float64)
                dr = np.divide(num, den, out=out_array, where=(den != 0))
            else:
                print(f"No financial data found for {ticker}")
                roe = np.append(roe, np.nan)
                om = np.append(om, np.nan)
                nm = np.append(nm, np.nan)
                dr = np.append(dr, np.nan)
        else:
            print(f"Failed to fetch data for {ticker}: {response.status_code}")
            
    new_line = {'Ticker':ticker, 'Ratio':'ROE'}
    for date, arr in zip(years, roe):
        new_line[date] = arr
    new_df = pd.DataFrame(new_line, index=[0])
    # Concatenate the new DataFrame with the existing DataFrame
    df_tot = pd.concat([df_tot, new_df], ignore_index=True)
    
    new_line = {'Ticker':ticker, 'Ratio':'Op Margin'}
    for date, arr in zip(years, om):
        new_line[date] = arr
    new_df = pd.DataFrame(new_line, index=[0])
    # Concatenate the new DataFrame with the existing DataFrame
    df_tot = pd.concat([df_tot, new_df], ignore_index=True)
    
    new_line = {'Ticker':ticker, 'Ratio':'Net Margin'}
    for date, arr in zip(years, nm):
        new_line[date] = arr
    new_df = pd.DataFrame(new_line, index=[0])
    # Concatenate the new DataFrame with the existing DataFrame
    df_tot = pd.concat([df_tot, new_df], ignore_index=True)
    
    new_line = {'Ticker':ticker, 'Ratio':'Debt Ratio'}
    for date, arr in zip(years, dr):
        new_line[date] = arr
    new_df = pd.DataFrame(new_line, index=[0])
    # Concatenate the new DataFrame with the existing DataFrame
    df_tot = pd.concat([df_tot, new_df], ignore_index=True)

    # Note: This is a basic template. You'll need to handle errors, pagination, rate limits, etc.
```
